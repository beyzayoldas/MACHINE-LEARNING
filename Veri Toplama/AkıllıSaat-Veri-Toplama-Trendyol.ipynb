```python
#Gerekli Kütüphanaleri tanımlayalım.
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
#Gerekli listeleri tanımlayalım.
url_list = []
prices_list = []
brands_list = []
propTitles = []
propValues = []

#Özellikleri çekelim.
for i in range(1, 30):  # 1-30 sayfa aralığındaki veriyi alıyoruz!!
    url = "https://www.trendyol.com/sr?q=ak%C4%B1ll%C4%B1%20saat&qt=ak%C4%B1ll%C4%B1%20saat&st=ak%C4%B1ll%C4%B1%20saat&os=1" + str(i)  # Urlin for döngüsü ile güncellenmesi
    r = requests.get(url)  # Hazırlanan url'e istek atılması
    source = BeautifulSoup(r.content, "html.parser",from_encoding='utf-8')  # İstek yapılan sayfanın içeriğinin çekilmesi

    urls = source.find_all("div", attrs={"class": "p-card-chldrn-cntnr"})  # Tüm verilerin bulunması
    for url in urls:
        url_saat = "https://www.trendyol.com/" + url.a.get("href")  # Verilerin linklerinin bulunması
        url_list.append(url_saat)  # Linklerin kaydedilmesi
        print(url_saat)

        r_saat = requests.get(url_saat)  # Bulunan linke tekrar istek atılması
        source_saat = BeautifulSoup(r_saat.content, "html.parser", from_encoding='utf-8')  # Veri içeriğnin çekilmesi

        properties = source_saat.find_all("li", attrs={"class": "detail-attr-item"})  # Bütün özelliklerin bulunması
        for prop in properties:
            span = prop.find_all("span")
            
            if len(span) == 2: #Eğer 2 span bulunmuşsa
                # İlk span'ı propTitles listesine ekle
                propTitles.append(span[0].text)
                # İkinci span'ı propTitles listesine ekle
                propValues.append(span[1].text)

    prices = source.find_all("div", attrs={"class": "prc-box-dscntd"})  # Bütün fiyat özelliklerin bulunması
    for price in prices:
         prices_list.append(price.text)  # Bulunan özellikler arasında gezinme ve listeye yazma
         print(price.text)
        
    brands = source.find_all("span", attrs={"class": "prdct-desc-cntnr-ttl"})  # Bütün marka özelliklerin bulunması
    for brand in brands:
         brands_list.append(brand.text)  # Bulunan özellikler arasında gezinme ve listeye yazma
         print(brand.text)
    

print(str(len(url_list)) + " Adet link bulundu.")
print(str(len(prices_list)) + " Adet fiyat bulundu")
print(str(len(brands_list)) + " Adet marka bulundu")
print(str(len(propTitles)) + " Adet özellik başlığı bulundu")
print(str(len(propValues)) + " Adet özellik verisi bulundu")
```

```python
#url ve fiyatları bir dataframe 'e yazma
df_urls = pd.DataFrame()
df_urls["brands"] = brands_list
df_urls["urls"] = url_list
df_urls["prices"] = prices_list
```

```python
#Bulunan veri sayısı
saat = len(url_list)
```

```python
#Bulunan özellik başlıklarının benzersizlerini bulma
columns = np.array(propTitles)
columns = np.unique(columns)
```

```python
#Başlıkları kullanarak url ve fiyat ile birlikte yeni bir dataframe oluşturma
df = pd.DataFrame(columns = columns)
df["brand"] = brands_list
df["url"] = url_list
df["price"] = prices_list
```

```python
#Oluşturulan dataframe gösterme
df.head()
```

```python
#Dataframe'i kullanarak bütün verileri çekme ve sütunlara yazma

for i in range(0, saat):
    url = df["url"].loc[i]
    r = requests.get(url)
    source = BeautifulSoup(r.content, "lxml", from_encoding='utf-8')
    
    properties = source.find_all("li", attrs={"class": "detail-attr-item"})  # Bütün özelliklerin bulunması
    for prop in properties:
        span = prop.find_all("span")
        
        if len(span) == 2:
            prop_name = span[0].text
            prop_value = span[1].text
            print(f" {prop_name} : {prop_value}")
            df[prop_name].loc[i] = prop_value
        
```
