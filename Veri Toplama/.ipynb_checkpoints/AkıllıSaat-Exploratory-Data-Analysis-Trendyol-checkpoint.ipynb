```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel(r'C:\Users\user\OneDrive - Manisa Celal Bayar Üniversitesi\Masaüstü\Beyza Yoldaş\Veri Toplama\Saat Veri Toplama.xlsx')
```

```python
data.info()
```

```python
#Aktarımda site tarafından verilmeyen bazı boş bilgileri filtreleme işlemi
data = data[data.apply(lambda row: row.count() > 3, axis=1)]
data
```

```python
data.head(18)
```

```python
plt.figure(figsize=(16,16))
plt.subplot(2,1,1)
sns.countplot(x ='brand',data=data,order = data['brand'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

```python
for i in range(0,100):
    data["price"] = data.price.str.replace(',{}'.format(i),'')

data["price"] = data.price.str.replace(',00'.format(i),'')
data["price"] = data.price.str.replace(',01'.format(i),'')
data["price"] = data.price.str.replace(',02'.format(i),'')
data["price"] = data.price.str.replace(',03'.format(i),'')
data["price"] = data.price.str.replace(',04'.format(i),'')
data["price"] = data.price.str.replace(',05'.format(i),'')
data["price"] = data.price.str.replace(',06'.format(i),'')
data["price"] = data.price.str.replace(',07'.format(i),'')
data["price"] = data.price.str.replace(',08'.format(i),'')
data["price"] = data.price.str.replace(',09'.format(i),'')
data["price"] = data.price.str.replace('.','')
```

```python
data["price"] = data.price.str.extract('(\d+\.\d+|\d+)').astype(float)
```

```python
data["price"]
```

```python
data["Kasa Çapı"]
```

```python
data = data.rename(columns={'Kasa Çapı':'Kasa_capi', 'Garanti Süresi':'Garanti_suresi', 'Ekran Boyut Aralığı':'Ekran_boyut_araligi', 'Batarya Kapasitesi Aralığı':'Batarya_kapasitesi_araligi'})
```

```python
data["Kasa_capi"]
```

```python
data["Kasa_capi"]=data.Kasa_capi.str.replace(' mm','')
```

```python
data["Kasa_capi"].replace('', np.nan, inplace=True)
```

```python
data["Kasa_capi"].replace('Belirtilmemiş', np.nan, inplace=True)
```

```python
data["Kasa_capi"]
```

```python
data["Kasa_capi"] = data["Kasa_capi"].apply(lambda x: 
    (float(x.split('-')[0].replace('"', '').strip()) + float(x.split('-')[1].replace('"', '').strip())) / 2 
    if isinstance(x, str) and '-' in x 
    else (float(x.replace('"', '').strip()) if isinstance(x, str) else float(x))
)
```

```python
data["Kasa_capi"]
```

```python
data["Garanti_suresi"]=data.Garanti_suresi.str.replace('Yıl','')
data["Garanti_suresi"]=data.Garanti_suresi.str.replace('6 Ay','0')
data["Garanti_suresi"]=data.Garanti_suresi.str.replace('6Ay','')
```

```python
data["Garanti_suresi"].dropna(how='all')
```

```python
data["Garanti_suresi"]
```

```python
data["Ekran_boyut_araligi"]
```

```python
# data["Ekran_boyutu"] = data["Ekran_boyutu"].str.replace(' inç', '').str.strip()
data["Ekran_boyut_araligi"] = data["Ekran_boyut_araligi"].str.replace("'''", '').str.strip()
data["Ekran_boyut_araligi"] = data["Ekran_boyut_araligi"].str.replace("'", '').str.strip()
```

```python
data["Ekran_boyut_araligi"]
```

```python
data["Ekran_boyut_araligi"]
```

```python
data["Ekran_boyut_araligi"] = data["Ekran_boyut_araligi"].apply(lambda x: 
    (float(x.split('-')[0].replace('"', '').strip()) + float(x.split('-')[1].replace('"', '').strip())) / 2 
    if isinstance(x, str) and '-' in x 
    else (float(x.replace('"', '').strip()) if isinstance(x, str) else float(x))
)
```

```python
data["Batarya_kapasitesi_araligi"]
```

```python
data["Batarya_kapasitesi_araligi"]=data.Batarya_kapasitesi_araligi.str.replace('mAh','')
data["Batarya_kapasitesi_araligi"]=data.Batarya_kapasitesi_araligi.str.replace('altı','')
data["Batarya_kapasitesi_araligi"]=data.Batarya_kapasitesi_araligi.str.replace('mAh altı','')
```

```python
data["Batarya_kapasitesi_araligi"]
```

```python
data["Batarya_kapasitesi_araligi"] = data["Batarya_kapasitesi_araligi"].apply(lambda x: (float(x.split('-')[0]) + float(x.split('-')[1])) / 2 if (isinstance(x, str) and '-' in x) else float(x))
```

```python
data["Kasa_capi"]=data.Kasa_capi.astype(float)
data["Garanti_suresi"]=data.Garanti_suresi.astype(float)
data["Ekran_boyut_araligi"]=data.Ekran_boyut_araligi.astype(float)
data["Batarya_kapasitesi_araligi"]=data.Batarya_kapasitesi_araligi.astype(float)
```

```python
plt.figure(figsize=(16,16))
plt.subplot(2,1,1)
sns.countplot(x='İşletim Tipi',data=data,order=data['İşletim Tipi'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

```python
print(f"Max Fiyat: {str(data['price'].max()).replace('0.0', '')} TL -->  {str(data['url'].loc[data['price'].idxmax()])}")
print(f"Max Kasa_capi: {str(data['Kasa_capi'].max())} mm -->  {str(data['url'].loc[data['Kasa_capi'].idxmax()])}")
print(f"Max Garanti_suresi: {str(data['Garanti_suresi'].max())} Yıl -->  {str(data['url'].loc[data['Garanti_suresi'].idxmax()])}")
print(f"Max Ekran_boyut_araligi: {str(data['Ekran_boyut_araligi'].max())} inç -->  {str(data['url'].loc[data['Ekran_boyut_araligi'].idxmax()])}")
print(f"Max Batarya_kapasitesi_araligi: {str(data['Batarya_kapasitesi_araligi'].max())} maH --> {str(data['url'].loc[data['Batarya_kapasitesi_araligi'].idxmax()])}")
```
